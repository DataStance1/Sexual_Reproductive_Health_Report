Marking Scheme

**Question a: Which of these best describes an unprotected sex?**
'Sex without condom' = 1
'Sex without a security guard'= 0
'Sex in an unsafe place'= 0

Question b: have you used condom before?
No right or wrong anwser

Question c: condoms make sex less enjoyable?
'strongly agree'= 0
'agree' = 0
'neutral' = 0
'disagree' = 1
'strongly disagree'= 2

Question d: Is it ok to use vaseline as lubricant with a latex condom?
False = 1
True = 0

Question e: Oral and anal sex are safe alternatives to vaginal sex
False = 1
True = 0

Question f: What does STDs stand for?
'Sexually Transmitted Diseases' = 1
'Sex Time Daily' = 0
'Sexually Transferrable Diseases' = 0

Question g: Even if a condom is used 100% correctly, it does not give complete protection to STDs 
True = 1
False = 0

Question h: To avoid aids the abc strategy is recommended. what does abc stand for?
'Abstinence, Be faithful, Condom use'= 1
'Alternate, Be wise, Condom use' = 0
'Always, Be consistent, Condom use' = 0

Question i: One can only get HIV through having unprotected vaginal sex with an infected person
False = 1
True = 0

Question j: Taking antibiotics such as ampiclox before and/or after sex reduces the chances of getting STDs
False = 1
True = 0


```python
#marking question a
score_a = []

for i in df_clean.a:
    if i == 'Sex without condom':
        score_a.append(1)
    else:
        score_a.append(0)
score_a
```

```python
# marking question c
score_c = []

for i in df_clean.c:
    if i == 'disagree':
        score_c.append(1)
    elif i == 'strongly disagree':
        score_c.append(2)
    else:
        score_c.append(0)
score_c
```

```python
# marking question d: Is it ok to use vaseline as lubricant with a latex condom?
score_d = []

for i in df_clean.d:
    if i == False:
        score_d.append(1)
    else:
        score_d.append(0)
score_d
```

```python
# marking question e: Oral and anal sex are safe alternatives to vaginal sex
score_e = []

for i in df_clean.e:
    if i == False:
        score_e.append(1)
    else:
        score_e.append(0)
score_e
```

```python
# marking question f: What does STDs stand for?
score_f = []

for i in df_clean.f:
    if i == 'Sexually Transmitted Diseases':
        score_f.append(1)
    else:
        score_f.append(0)
score_f
```

```python
# marking question g: Even if a condom is used 100% correctly, it does not give complete protection to STDs
score_g = []

for i in df_clean.g:
    if i == True:
        score_g.append(1)
    else:
        score_g.append(0)
score_g 
```

```python
# marking question h: To avoid aids the abc strategy is recommended. what does abc stand for?
score_h = []

for i in df_clean.h:
    if i == 'Abstinence, Be faithful, Condom use':
        score_h.append(1)
    else:
        score_h.append(0)
score_h
```

```python
# marking question i: One can only get HIV through having unprotected vaginal sex with an infected person
score_i = []

for i in df_clean.i:
    if i == 'False':
        score_i.append(1)
    else:
        score_i.append(0)
score_i
```

```python
df_clean.i.unique()
```

```python
#marking question j:Taking antibiotics such as ampiclox before and/or after sex reduces the chances of getting STDs
score_j = []

for i in df_clean.j:
    if i == False:
        score_j.append(1)
    else:
        score_j.append(0)
score_j
```

```python
df_clean['score_a'] = score_a
df_clean['score_c'] = score_c
df_clean['score_d'] = score_d
df_clean['score_e'] = score_e
df_clean['score_f'] = score_f
df_clean['score_g'] = score_g
df_clean['score_h'] = score_h
df_clean['score_i'] = score_i
df_clean['score_j'] = score_j

df_clean
```

```python
df_clean['total_score'] = df_clean[['score_a', 'score_c', 'score_d', 'score_e', 'score_f', 'score_g', 'score_h', 'score_i', 'score_j']].sum(axis=1)
df_clean
```

```python
# categorizing the respondents based on their total score- those with score greater than the average total
# score have good knowledge while those with scores less than the average total score have poor knowledge
knowledge_category = []

for i in df_clean.total_score:
    if i >= round(df_clean.total_score.mean()):
        knowledge_category.append("good knowledge")
    else:
        knowledge_category.append("poor knowledge")
knowledge_category
```

```python
df_clean["knowledge_category"] = knowledge_category
df_clean
```
